```python
import gradio as gr
import requests
import openai

# Set up OpenAI API key
openai.api_key = 'api-key'

# Crop data with specific requirements
crops_data = {
    'wheat': {'min_temp': 10, 'max_temp': 25, 'min_moisture': 30},
    'rice': {'min_temp': 20, 'max_temp': 35, 'min_moisture': 50},
    'cotton': {'min_temp': 15, 'max_temp': 30, 'min_moisture': 40},
    'maize': {'min_temp': 18, 'max_temp': 32, 'min_moisture': 50},
    'barley': {'min_temp': 5, 'max_temp': 20, 'min_moisture': 40},
    'sorghum': {'min_temp': 20, 'max_temp': 35, 'min_moisture': 45},
    'soybean': {'min_temp': 20, 'max_temp': 30, 'min_moisture': 50},
    'sunflower': {'min_temp': 18, 'max_temp': 30, 'min_moisture': 40},
    'peanut': {'min_temp': 25, 'max_temp': 40, 'min_moisture': 35},
    'mustard': {'min_temp': 10, 'max_temp': 25, 'min_moisture': 35}
}


# Separate weather data for Rain and No Rain conditions
zone_weather_data_rain = {
    'AB1 Block': {'temperature': 29, 'humidity': 76},
    'AB2 Block': {'temperature': 26, 'humidity': 72},
    'A-Block Hostel': {'temperature': 25, 'humidity': 73},
    'North Sqaure': {'temperature': 29, 'humidity': 76},
    'Library Area': {'temperature': 27, 'humidity': 76},
    'Football Area': {'temperature': 27, 'humidity': 81},
    'Gazebo Area': {'temperature': 28, 'humidity': 65},
    'Clock Tower': {'temperature': 30, 'humidity': 78},
    'Aavin': {'temperature': 28, 'humidity': 74},
    'C-Block Hostel': {'temperature': 27, 'humidity': 77},
    # Add more zones and their respective weather data for rain
}

zone_weather_data_no_rain = {
    'AB1 Block': {'temperature': 33, 'humidity': 60},
    'AB2 Block': {'temperature': 35, 'humidity': 62},
    'A-Block Hostel': {'temperature': 32, 'humidity': 61},
    'North Sqaure': {'temperature': 34, 'humidity': 65},
    'Library Area': {'temperature': 37, 'humidity': 58},
    'Football Area': {'temperature': 38, 'humidity': 59},
    'Gazebo Area': {'temperature': 32, 'humidity': 58},
    'Clock Tower': {'temperature': 36, 'humidity': 63},
    'Aavin': {'temperature': 32, 'humidity': 62},
    'C-Block Hostel': {'temperature': 37, 'humidity': 57},
    # Add more zones and their respective weather data for no rain
}

# Sample fertilizer data
fertilizer_data = {
    'wheat': {'type': 'NPK 10-20-10', 'quantity_per_hectare': '100 kg', 'timing': 'Pre-sowing'},
    'rice': {'type': 'Urea', 'quantity_per_hectare': '150 kg', 'timing': 'Top dressing at tillering stage'},
    'cotton': {'type': 'Phosphorous-based', 'quantity_per_hectare': '75 kg', 'timing': 'Early growth stage'},
    'maize': {'type': 'NPK 20-10-10', 'quantity_per_hectare': '120 kg', 'timing': 'Pre-sowing and at knee-high stage'},
    'barley': {'type': 'Urea', 'quantity_per_hectare': '100 kg', 'timing': 'At sowing and at tillering stage'},
    'sorghum': {'type': 'NPK 15-15-15', 'quantity_per_hectare': '80 kg', 'timing': 'Pre-sowing'},
    'soybean': {'type': 'Lime and Phosphorous', 'quantity_per_hectare': '40 kg', 'timing': 'At sowing'},
    'sunflower': {'type': 'NPK 12-24-12', 'quantity_per_hectare': '90 kg', 'timing': 'At sowing and at flowering stage'},
    'peanut': {'type': 'Phosphorous-based', 'quantity_per_hectare': '50 kg', 'timing': 'At sowing and flowering stage'},
    'mustard': {'type': 'NPK 20-20-0', 'quantity_per_hectare': '80 kg', 'timing': 'At sowing and at pod formation'}
}



def chat_with_gpt(prompt, chat_history=[]):
    chat_history.append({"role": "user", "content": prompt})
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=chat_history
    )
    reply = response.choices[0].message.content.strip()
    chat_history.append({"role": "assistant", "content": reply})
    return reply, chat_history


def interactive_chatbot(user_input, chat_history=[]):
    if user_input.lower() in ["quit", "exit", "bye"]:
        return "Goodbye! Feel free to return any time for assistance.", chat_history

    response, chat_history = chat_with_gpt(user_input, chat_history)

    # Adding suggestions for the user to keep the conversation engaging
    suggestions = "You can ask about crop conditions, weather recommendations, or other farming tips!"
    interactive_response = f"{response}\n\n**Suggestions**: {suggestions}"

    return interactive_response, chat_history

# Function to check if it’s safe to grow the selected crop
def check_crop_safety(crop, temperature, humidity):
    crop_info = crops_data.get(crop.lower())
    if not crop_info:
        return "Crop data not available."

    # Check if the temperature and humidity are suitable
    if crop_info['min_temp'] <= temperature <= crop_info['max_temp'] and humidity >= crop_info['min_moisture']:
        return f"It is safe to grow {crop.capitalize()}."
    else:
        return f"It is not safe to grow {crop.capitalize()} under current conditions."

# Main function for fetching crop safety based on campus zone, crop name, and rain condition
def fetch_crop_safety(zone, crop, is_raining):
    if is_raining == "Rain":
        weather_data = zone_weather_data_rain.get(zone)
    else:
        weather_data = zone_weather_data_no_rain.get(zone)

    if not weather_data:
        return "Zone not found or weather data not available."

    temperature = weather_data['temperature']
    humidity = weather_data['humidity']

    crop_safety_message = check_crop_safety(crop, temperature, humidity)
    return f"Zone: {zone}\nTemperature: {temperature}°C\nHumidity: {humidity}%\nRain: {is_raining}\n\n{crop_safety_message}"





# Function to get fertilizer recommendation
def get_fertilizer_recommendation(crop):
    if crop in fertilizer_data:
        recommendation = fertilizer_data[crop]
        return (f"For {crop.capitalize()}, use {recommendation['type']} fertilizer."
                f"\nRecommended quantity: {recommendation['quantity_per_hectare']} per hectare."
                f"\nApplication timing: {recommendation['timing']}")
    else:
        return "Fertilizer information is not available for this crop."



def upload_file_to_s3(file_path, bucket_name, object_name=None):
    try:
        if object_name is None:
            object_name = file_path

        s3.upload_file(file_path, bucket_name, object_name)
        print(f"File {file_path} uploaded successfully to {bucket_name}/{object_name}")
    except NoCredentialsError:
        print("Credentials not available.")


def get_pdf_url(bucket_name, file_name):
    # Correctly format the URL
    s3_url = f"https://{bucket_name}.s3.amazonaws.com/{file_name}"
    return s3_url

def fetch_learning_resource(crop):
    bucket_name = "learning-resources-bucket"

    file_name = f"{crop}.pdf"

    resource_url = get_pdf_url(bucket_name, file_name)

    return f"Here is the learning resource for {crop}: [Download PDF]({resource_url})"


# Add Gradio interface for fertilizer recommendation
fertilizer_interface = gr.Interface(
    fn=get_fertilizer_recommendation,
    inputs=gr.Dropdown(label="Select Crop", choices=list(fertilizer_data.keys())),
    outputs="text",
    title="Fertilizer Recommendation",
    description="Select a crop to receive fertilizer recommendations."
)

# Gradio interface setup
crop_safety_interface = gr.Interface(
    fn=fetch_crop_safety,
    inputs=[
        gr.Dropdown(label="Select Campus Zone", choices=list(zone_weather_data_no_rain.keys()) + list(zone_weather_data_rain.keys())),
        gr.Dropdown(label="Select Crop", choices=list(crops_data.keys())),
        gr.Dropdown(label="Is it raining?", choices=["Rain", "No Rain"])  # Added "Rain" or "No Rain" input
    ],
    outputs="text",
    title="VIT Chennai Crop Management - Crop Safety Checker",
    description="Choose a campus zone and crop to check if the weather is suitable for growing that crop, considering rain conditions."
)



chatbot_interface = gr.Interface(
    fn=interactive_chatbot,
    inputs=[
        gr.Textbox(label="You", placeholder="Ask me something about crop safety or farming tips."),
        gr.State([])  # To hold the chat history
    ],
    outputs=[
        gr.Textbox(label="Chatbot"),
        gr.State([])  # Return updated chat history
    ],
    title="Interactive Farming Chatbot",
    description="An interactive chatbot for crop and weather-related queries. Keep the conversation going with helpful suggestions!"
)


learning_resources_interface = gr.Interface(
    fn=fetch_learning_resource,
    inputs=gr.Dropdown(choices=["Wheat", "Rice", "Cotton"], label="Select Crop"),
    outputs=gr.Markdown(),
    title="Learning Resources for Crops",
    description="Select a crop to get the learning resource PDF."
)

# Gradio interface for feedback with a link to Google Form
feedback_link = "https://forms.gle/UrrUcV3EJTX9okm36"

support_interface = gr.Interface(
    fn=lambda: f"Please provide your feedback [here]({feedback_link}).",
    inputs=[],
    outputs="markdown",
    title="Feedback Support",
    description="We value your feedback! Please let us know your thoughts or suggestions by filling out the form."
)



# Dark mode CSS with background image
css = """

h1, h2 {
  color: #67dbca; /* Lighter primary color */
}

.container {
    border-radius: 8px;
    background: linear-gradient(135deg, rgba(42, 42, 42, 0.8), rgba(10, 10, 10, 0.8));
    padding: 20px;
    box-shadow: 0px 4px 10px rgba(0, 0, 0, 0.3);
    transition: box-shadow 0.3s ease;
}

.container:hover {
    box-shadow: 0px 8px 20px rgba(0, 0, 0, 0.5);
}

button {
  background: linear-gradient(135deg, #3498db, #03a9f4); /* Gradient for buttons */
  color: #fff;
}

button:hover {
    background: linear-gradient(135deg, #03a9f4, #3498db);
}

textarea {
    resize: vertical;
    width: 100%;
    padding: 10px;
    border: 1px solid #555;
    border-radius: 5px;
    background: linear-gradient(135deg, #333, #222); /* Gradient background for textarea */
    color: #f0f0f0;
    transition: border-color 0.3s ease;
}

textarea:focus {
    border-color: #4fc3f7;
    outline: none;
}
"""

app = gr.TabbedInterface(
    [crop_safety_interface, chatbot_interface, support_interface, fertilizer_interface, learning_resources_interface],
    ["Crop Safety Checker 🌾", "Help 💁🏻", "Support 💭", "Fertilizer Recommendations 🌱", "Learning Resources 📚"],
    css=css
)

app.launch()
```
